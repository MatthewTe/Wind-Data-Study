```python
# Importing frameworks: 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Cassia_Compiled_Wind_Data.csv",parse_dates=True, index_col="datetime")
```

```python
# Confirming that the data frame is sucessfully imported: 
df.head()
```

```python
# Visualizing the un-cleaned data:
df.plot()
# Note how there is far to much data in the data frame to be accurately visualized: 
```

```python
# creating a rolling average to collect a data point a day: 
df = df.rolling(window=4000).mean()
```

```python
# Plotting the recorded windspeed for the period of 2007 to 2011:
df["WindSpd_m/s"].plot()
plt.xlabel("Time")
plt.ylabel("Wind speed m/s")
plt.title("Average Cassia Platform Windspeed")
```

```python
# Plotting the wind power that would be generated from the WindSpd data: 
df["Power_Output/MWt"].plot()
plt.xlabel("Time")
plt.ylabel("Power Output MWt")
plt.title("Average Cassia Platform Single Turbine Output")
```

```python
# Plotting the recorded windspeed and the single turbine output on one graph for comparrison: 
df.plot()
plt.title("Comparison between the actual Wind Speed and the Single Turbine Output")
plt.xlabel("Time")
```

```python
# Importing frameworks: 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Cassia_Compiled_Wind_Data.csv",parse_dates=True)
```

```python
# Confirming that the data frame is sucessfully imported: 
df.head()
```

```python
plt.plot(df["Power_Output/MWt"])
plt.show()
```

```python
# Creating a moving average every 1000 data points:  
df_roll = df["Power_Output/MWt"].rolling(1000).mean()
```

```python
# plotting the new data frame that contains the rolling mean:
df_roll.plot()
plt.show()
```
